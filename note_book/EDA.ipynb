```python

train_data = pd.read_csv('train.csv')
val_data = pd.read_csv('val.csv')
test_data = pd.read_csv('test.csv')
```

```python


def check_duplicates(train_df, val_df, test_df, filename_col="filename"):

    # Lấy danh sách filename từ mỗi tập dưới dạng tập hợp
    train_files = set(train_df[filename_col])
    val_files = set(val_df[filename_col])
    test_files = set(test_df[filename_col])

    # Tìm sự trùng lặp giữa các tập
    train_val_dup = train_files.intersection(val_files)
    train_test_dup = train_files.intersection(test_files)
    val_test_dup = val_files.intersection(test_files)

    # In kết quả
    print(f"Số lượng trùng lặp giữa Train và Val: {len(train_val_dup)}")
    if train_val_dup:
        print("Danh sách trùng (Train & Val):", train_val_dup)
    print(f"Số lượng trùng lặp giữa Train và Test: {len(train_test_dup)}")
    if train_test_dup:
        print("Danh sách trùng (Train & Test):", train_test_dup)
    print(f"Số lượng trùng lặp giữa Val và Test: {len(val_test_dup)}")
    if val_test_dup:
        print("Danh sách trùng (Val & Test):", val_test_dup)

# Ví dụ sử dụng
check_duplicates(train_data, val_data, test_data, filename_col="filename")
```

```python
print("Train data shape:", train_data.shape)
print("Validation data shape:", val_data.shape) 
print("Test data shape:", test_data.shape)
```

```python
 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
train_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax1)
ax1.set_title('Train')

val_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax2)
ax2.set_title('Validation')

test_data['score'].value_counts(normalize=True).plot(kind='bar', ax=ax3)
ax3.set_title('Test')

plt.tight_layout()
plt.show()
```
